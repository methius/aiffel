```python
# 필요 라이브러리 import

import os
import glob

import tensorflow as tf
from tensorflow import keras
import numpy as np

import matplotlib.pyplot as plt
from PIL import Image
```

```python
image_dir_path_train = '/Users/joowanha/PycharmProjects/AIFFEL/rock_scissors_paper/train'
image_dir_path_test = '/Users/joowanha/PycharmProjects/AIFFEL/rock_scissors_paper/test'
```

```python
# 이미지 사이즈 변환
def resize_images(img_path):
    images=glob.glob(img_path + "/*.jpg")
    
    print(len(images), " images to be resized.")
    
    # 파일마다 모두 28x28 사이즈로 바꾸어 저장합니다.
    target_size = (28, 28)
    for img in images :
        old_img=Image.open(img)
        new_img=old_img.resize(target_size,  Image.Resampling.LANCZOS)
        new_img.save(img, "JPEG")
    
    print(len(images), " images resized.")
```

```python
# training 이미지가 저장된 디렉토리 아래의 모든 jpg 파일을 읽어 들여서 변환
print("가위 이미지변환")
resize_images(image_dir_path_train + '/scissors')

print("바위 이미지변환")
resize_images(image_dir_path_train+'/rock')

print("보 이미지변환")
resize_images(image_dir_path_train +'/paper')
```

```python
# test 이미지 변환

print("가위 이미지변환")
resize_images(image_dir_path_test + '/scissors')

print("바위 이미지변환")
resize_images(image_dir_path_test+'/rock')

print("보 이미지변환")
resize_images(image_dir_path_test +'/paper')
```

```python
# load_data 함수
def load_data(img_path, number_of_data):
    # 가위 : 0, 바위 : 1, 보 : 2
    img_size = 28
    color =3
    
    print(number_of_data)
    # 이미지 데이터와 라벨(가위 : 0, 바위 : 1, 보 : 2) 데이터를 담을 행렬(matrix) 영역을 생성합니다.
    imgs = np.zeros(number_of_data * img_size * img_size * color, dtype=np.int32).reshape(number_of_data,img_size,img_size,color)
    labels = np.zeros(number_of_data, dtype = np.int32)
    
    idx = 0
    for file in glob.iglob(img_path+'/scissors/*.jpg'):
        img = np.array(Image.open(file), dtype = np.int32)
        imgs[idx, :, :, : ] = img # 데이터 영역에 이미지 행렬을 복사
        labels[idx] = 0 # 가위
        idx = idx +1
        
    for file in glob.iglob(img_path+'/rock/*.jpg'):
        img = np.array(Image.open(file), dtype = np.int32)
        imgs[idx, :, :, : ] = img # 데이터 영역에 이미지 행렬을 복사
        labels[idx] = 1 # 바위
        idx = idx +1           
        
    for file in glob.iglob(img_path+'/paper/*.jpg'):
        img = np.array(Image.open(file), dtype = np.int32)
        imgs[idx, :, :, : ] = img # 데이터 영역에 이미지 행렬을 복사
        labels[idx] = 2 # 보
        idx = idx +1   
        
    print("데이터의 이미지 개수는 ", idx, "입니다.")
    return imgs, labels
```

```python
tr_scissors = len(os.listdir(image_dir_path_train+'/scissors'))
tr_rock = len(os.listdir(image_dir_path_train+'/rock'))
tr_paper =len(os.listdir(image_dir_path_train+'/paper'))
train_total = tr_scissors + tr_rock +tr_paper

train_total
```

```python
x_train, y_train = load_data(image_dir_path_train, train_total)
```

```python
x_train.shape
```

```python
ts_scissors = len(os.listdir(image_dir_path_test+'/scissors'))
ts_rock = len(os.listdir(image_dir_path_test+'/rock'))
ts_paper =len(os.listdir(image_dir_path_test+'/paper'))
test_total = ts_scissors + ts_rock +ts_paper

test_total
```

```python
x_test, y_test = load_data(image_dir_path_test, test_total)
```

```python
x_train.max(), x_train.min()
```

```python
x_train_norm = x_train/255.0 # 0~1 사이의 값으로 정규화
x_test_norm = x_test/255.0
```

```python
print('x_train shape: {}'.format(x_train_norm.shape))
print('y_train shape: {}'.format(y_train.shape)) 
print('x_test shape: {}'.format(x_test_norm.shape))
print('y_test shape: {}'.format(y_test.shape)) 
```

```python
plt.imshow(x_train[0])
print('라벨: ', y_train[0])
```

## 딥러닝 네트워크 설계하기

### [1차] 기본 baseline 적용

```python
x_train_reshaped=x_train_norm.reshape( -1, 28, 28, 3)  # 데이터갯수에 -1을 쓰면 reshape시 자동계산됩니다.
x_test_reshaped=x_test_norm.reshape( -1, 28, 28, 3)

print(f"After Reshape - x_train_reshaped shape: {x_train_reshaped.shape}")
print(f"After Reshape - x_test_reshaped shape: {x_test_reshaped.shape}")
```

```python
n_channel_1 = 16
n_channel_2 = 32
epochs_num = 10

model_base=keras.models.Sequential()
model_base.add(keras.layers.Conv2D(n_channel_1, (3,3), activation='relu', input_shape=(28,28,3)))
model_base.add(keras.layers.MaxPool2D(2,2))
model_base.add(keras.layers.Conv2D(n_channel_2, (3,3), activation='relu'))
model_base.add(keras.layers.MaxPooling2D((2,2)))
model_base.add(keras.layers.Flatten())
model_base.add(keras.layers.Dense(32, activation='relu'))
model_base.add(keras.layers.Dense(3, activation='softmax'))

print('Model에 추가된 Layer 개수: ', len(model_base.layers))

model_base.summary()
```

```python
model_base.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])

# model.fit(x_train_reshaped, y_train, epochs=10)
history_base = model_base.fit(x_train_reshaped, y_train, epochs=epochs_num, validation_data=(x_test_reshaped, y_test))


test_loss, test_accuracy = model_base.evaluate(x_test_reshaped,y_test, verbose=2)
print(f"test_loss: {test_loss}")
print(f"test_accuracy: {test_accuracy}")

predicted_result = model_base.predict(x_test_reshaped)  # model이 추론한 확률값.
predicted_labels = np.argmax(predicted_result, axis=1)

idx=0  #1번째 x_test를 살펴보자.
print('model.predict() 결과 : ', predicted_result[idx])
print('model이 추론한 가장 가능성이 높은 결과 : ', predicted_labels[idx])
print('실제 데이터의 라벨 : ', y_test[idx])
```

```python
plt.figure()

for k,v in history_base.history.items():
    plt.plot(history_base.history[k], label=k)
plt.legend()
plt.show()
```

### [2차] 13개 layer 적용

```python
n_channel_1 = 100
n_channel_2 = 100
n_dense=100
epochs_num = 20

model_1st = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(n_channel_1, (3, 3), activation='relu', input_shape=(28, 28, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Conv2D(n_channel_2, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(3, activation='softmax')
])

model_1st.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])

print('Model에 추가된 Layer 개수: ', len(model_1st.layers))

history_1st = model_1st.fit(x_train_reshaped, y_train, epochs=epochs_num, validation_data=(x_test_reshaped, y_test))

test_loss, test_accuracy = model_1st.evaluate(x_test_reshaped,y_test, verbose=2)
print(f"test_loss: {test_loss}")
print(f"test_accuracy: {test_accuracy}")
```

```python
plt.figure()

for k,v in history_1st.history.items():
    plt.plot(history_1st.history[k], label=k)
plt.legend()
plt.show()
```

### [3차] 15 layer 및 데이터 증강(rotation/flip) 적용

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
n_channel_1 = 300
n_channel_2 = 300
n_dense = 300
epochs_num = 10

print("테스트 개수 : ", len(x_train))

model_2nd = tf.keras.models.Sequential([
    tf.keras.layers.RandomFlip(mode="horizontal_and_vertical"),
    tf.keras.layers.RandomRotation(factor = 0.2),
    tf.keras.layers.Conv2D(n_channel_1, (3, 3), activation='relu', input_shape=(28, 28, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Conv2D(n_channel_2, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(3, activation='softmax')
])

model_2nd.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])

print('Model에 추가된 Layer 개수: ', len(model_2nd.layers))

history_2nd = model_2nd.fit(x_train_reshaped, y_train, epochs=epochs_num, validation_data=(x_test_reshaped, y_test))

test_loss, test_accuracy = model_2nd.evaluate(x_test_reshaped,y_test, verbose=2)
print(f"test_loss: {test_loss}")
print(f"test_accuracy: {test_accuracy}")
```

```python
plt.figure()

for k,v in history_2nd.history.items():
    plt.plot(history_2nd.history[k], label=k)
plt.legend()
plt.show()
```

### [4차] epochs 수 증대, learning rate reduction 적용

```python
from keras.callbacks import ReduceLROnPlateau

n_channel_1 = 200
n_channel_2 = 200
n_dense= 64
epochs_num = 50

model_3rd = tf.keras.models.Sequential([
    tf.keras.layers.RandomFlip(mode="horizontal_and_vertical"),
    tf.keras.layers.RandomRotation(factor = 0.2),
    # tf.keras.layers.RandomBrightness(factor = 0.2),
    # tf.keras.layers.BatchNormalization(),  
    tf.keras.layers.Conv2D(n_channel_1, (3, 3), activation='relu', input_shape=(28, 28, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Conv2D(n_channel_2, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(n_dense, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(3, activation='softmax')
])


learning_rate_reduction = ReduceLROnPlateau(monitor='val_accuracy',
                                            patience=2,
                                            verbose=1,
                                            factor=0.5,
                                            min_lr=0.000003)

model_3rd.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])

print('Model에 추가된 Layer 개수: ', len(model_3rd.layers))

history_3rd = model_3rd.fit(x_train_reshaped, y_train, epochs=epochs_num, validation_data=(x_test_reshaped, y_test), callbacks=[learning_rate_reduction])

test_loss, test_accuracy = model_3rd.evaluate(x_test_reshaped,y_test, verbose=2)
print(f"test_loss: {test_loss}")
print(f"test_accuracy: {test_accuracy}")
```

```python
plt.figure()

for k,v in history_3rd.history.items():
    plt.plot(history_3rd.history[k], label=k)
plt.legend()
plt.show()
```

## 회고  
 - 기본적으로 train data를 많이 확보하여 1차 baseline으로 적용했을 경우 60%가 넘는 정확도를 확보할 수 있었습니다.   
 - 정확도를 좀 더 높이기 위해 2차의 경우 모델 레이어 수를 늘렸으나 오히려 정확도는 감소했습니다. 단순히 레이어만 증대시키는 것이 바람직한 튜닝 방법은 아닐 수도 있겠다는 생각을 했습니다.   
 - 테스트 데이터의 자료를 사전에 확인한 결과 train 데이터와는 가위 바위 보 이미지의 방향이 다르다는 것을 알게되어 3차의 경우 회전과 플립을 적용했습니다. 2차의 결과보다는 나은 결과치가 나왔지만 여전히 baseline보다는 낮은 값이 나왔습니다.   
 - 4차의 경우 epoch 수를 키워서 테스트 해 본 결과, 정확도가 크게 올라가는 것을 확인하고 적정 epoch 수가 미치는 영향이 오히려 크다는 것을 확인하게 되었습니다. 하지만 train의 정확도보다 test의 정확도가 더 높게 나온 부분은 오히려 의문으로 남게 되었습니다.  
 - hyper-parameter가 테스트 결과에 미치는 영향을 좀 더 알아보기 위해 5차에서는 keras tuner를 활용하여 최적 파라미터 값을 찾아보는 테스트를 수행해 봤습니다. 그런데 searching 영역을 너무 넓게 둔 영향으로 10시간이 넘는 시간이 소요되었습니다. 시간(resource) 절약도 모델링의 중요 요소 중에 하나이기 때문에 parameter tunning할 때 좀 더 주의를 해야 할 것 같습니다. 그런데 아쉽게도 10시간 동안 seaching하여 찾은 파라미터 값도 loss가 너무 크게 나왔으며 4차에서 확보된 결과보다는 낮은 정확도를 보여주었습니다. 추후 기회가 될 때 다양한 테스트를 통해 tunning 방법에 대해 좀 더 확인해 보는 것이 좋을 것 같습니다.
