## (5) 모델 준비하기

```python
x_data[0]
```

```python
W = np.random.rand(len(x_data[0]))
b = np.random.rand()
```

```python
W
```

```python
b
```

```python
def model(X, W, b):
    predictions = 0
    for i in range(len(X[0])):
        predictions += X[:, i] * W[i]
    predictions += b
    return predictions
```

## (6) 손실함수 Loss 정의하기

```python
def MSE(a, b):
    mse = ((a-b) ** 2).mean()
    return mse
```

```python
def RMSE(mse):
    return mse**0.5
```

```python
def loss(X, W, b, y):
    predictions = model(X, W, b)
    L = MSE(predictions, y)
    return L
```

## (7) 기울기를 구하는 gradient 함수 구현하기

```python
X_train_diabetes.shape
```

```python
y_train_diabetes
```

```python
def gradient(X, W, b, y):
    N = len(W)
    y_pred = model(X, W, b)
    dW = 1/N * 2 * X.T.dot(y_pred - y)
    db = 2 * (y_pred - y).mean()
    return dW, db
```

```python
dW, db = gradient(X_train_diabetes, W, b, y_train_diabetes)
print("dW:", dW , "dW.shape:" , dW.shape)
print("db:", db)
```

## (8) 하이퍼 파라미터인 학습률 설정하기

```python
LEARNING_RATE = 0.05 # 0.001 , 0.01, 0.1 테스트 수행함
```

## (9) 모델 학습하기

```python
def finding_param(LEARNING_RATE, iteration, prt_step):
    losses = []
    W = np.random.rand(len(x_data[0]))
    b = np.random.rand()
    for i in range(1, iteration+1):
        dW, db = gradient(X_train_diabetes, W, b, y_train_diabetes)
        W -= LEARNING_RATE * dW
        b -= LEARNING_RATE * db
        L = loss(X_train_diabetes, W, b, y_train_diabetes)
        losses.append(L)
        if i % prt_step == 0 :
            print('Iteration %d : Loss %0.4f' % (i, L))
    print("W: ", W, "\n" , "b: ", b) 
    return (W, b, losses)
```

Loss가 수렴할 때까지 다양한 LEARNING_RATE와 iteration 횟수로 돌려보겠습니다.

```python
W, b, losses = finding_param(0.001,3000, 100)
```

```python
W_1000, b_1000, losses = finding_param(0.01,1000, 100)
```

```python
W, b, losses = finding_param(0.1,80000, 1000)
```

```python
W, b, losses = finding_param(0.05,150000, 1000)
```

LEARNING_RATE 0.05 로 150000회 정도 iteration 수행결과 140000회 정도에서 Loss가 2916에 수렴했습니다. 초기에는 Loss가 급격히 줄어들다가 일정 수준 이상 iteration 수행하면 감소폭이 현저히 감소하는 것을 확인할 수 있습니다.

Loss가 어느 정도 속도로 떨어지는 그래프로도 확인해 보겠습니다.

그런데 계산 결과 loss값이 최종 수렴하지 못한 W, b의 값에 대한 test 데이터 loss값이 더 작게 나왔다. overfitting 된 것일까? 이에 대한 답은 추후에 찾아보도록 하겠습니다.

비교를 위해 training data를 나누지 않고 전체 data를 기준으로 parameter와 loss를 구해보겠습니다.

```python
def finding_param_all(LEARNING_RATE, iteration, prt_step):
    losses = []
    W = np.random.rand(len(x_data[0]))
    b = np.random.rand()
    for i in range(1, iteration+1):
        dW, db = gradient(x_data, W, b, y_data)
        W -= LEARNING_RATE * dW
        b -= LEARNING_RATE * db
        L = loss(x_data, W, b, y_data)
        losses.append(L)
        if i % prt_step == 0 :
            print('Iteration %d : Loss %0.4f' % (i, L))
    print("W: ", W, "\n" , "b: ", b) 
    return (W, b)
```

```python
W_all, b_all = finding_param_all(0.05, 110000, 1000)
```

loss 값은 training data와 test data를 나누어서 학습한 결과보다는 적게 나왔습니다.

```python
W, b
```

```python
W_1000, b_1000
```

W_all, b_all 값을 W, b와 같이 비교해 볼 경우 b의 값은 유사한 값은 가지나 나머지 값들의 영향도를 명확히 이해하지 못하는 한 이 데이터로 추가 판단하기는 어려운 것 같습니다. 그렇다면 적합한 W, b를 구했다는 것을 어떻게 판단할 수 있을까요? 해당 방법에 대한 의문은 추후 해결할 수 있기를 기대하며 일단 물음표로 남겨둬야 겠습니다.

이번엔 target과 상관도가 높은 항목들만 모은 X 데이터들로 training 및 예측을 해보도록 하겠습니다.

```python
def finding_param_reduced(LEARNING_RATE, iteration, prt_step):
    losses = []
    W = np.random.rand(len(x_data_reduced[0]))
    b = np.random.rand()
    for i in range(1, iteration+1):
        dW, db = gradient(X_train_diabetes_red, W, b, y_train_diabetes_red)
        W -= LEARNING_RATE * dW
        b -= LEARNING_RATE * db
        L = loss(X_train_diabetes_red, W, b, y_train_diabetes_red)
        losses.append(L)
        if i % prt_step == 0 :
            print('Iteration %d : Loss %0.4f' % (i, L))
    print("W: ", W, "\n" , "b: ", b) 
    return (W, b)
```

```python
W_red, b_red = finding_param_reduced(0.005, 13000, 100)
```

```python
loss(X_test_diabetes_red, W_red, b_red, y_test_diabetes_red)
```

오히려 칼럼수를 줄였더니, loss 값은 더욱 커졌습니다. 상관도가 높은 항목들로만 구성하여 칼럼수를 줄이는 것은 loss 값을 줄이는데 항상 긍정적인 영향을 미치는 것은 아닌 것 같습니다.

MSE 또는 RMSE는 절대값 기준으로 결과가 나오기 때문에 실제 그 오차의 정도를 가늠하기가 어려운 부분이 있습니다. 이에 R squared 방법으로도 해당 모델의 성능을 평가하고 비교해 보도록 하겠습니다. 

```python
def cal_R2(X_test, W, b, y_test):
    y_pred = model(X_test, W, b)
    SST = np.sum((y_test - np.mean(y_test))**2)
    SSR = np.sum((y_test - y_pred)**2)
    R2 = 1 - (SSR/SST)
    return R2
```

```python
R2 = cal_R2(X_test_diabetes, W, b , y_test_diabetes)
```

```python
R2
```

R-Squared 값은 기본적으로 1에 가까울 수록 예측 성능이 매우 좋은 것으로 판단할 수 있으며, 일반적으로 사회과학 분야에서는 0.5이상도 상대적으로 의미있는 예측 정확도를 의미하기도 한다고 합니다. 0.7이상의 높은 값은 아니지만 상대적으로 의미있는 수준에서 예측 결과가 나왔다고 판단할 수 있을 것 같습니다.

gradient descent 알고리즘을 활용하여 손실함수를 직접 계산하여 최적 가중치와 편향값을 구해보고 이를 활용하여 당뇨 진행 정도에 대해 예측해보는 작업을 수행했습니다.

이에 추가로 R-Squared 값을 산출해 본 결과 0.5 수준이 나왔으며 매우 높은 수치는 아니지만 상대적으로 유의미한 결과치를 산출했다고 판단할 수 있을 것 같습니다.

holiday - whether the day is considered a holiday (휴일인 경우 숫자 1로 들어가 있는 것 같습니다. 추후 확인해 보겠습니다.)  
workingday - whether the day is neither a weekend nor holiday (주말 또는 휴일이 아닌 경우 1이 들어가 있는 것 같습니다. 해당 내용도 추후 확인해 보도록 하겠습니다.)  
weather - 1: Clear, Few clouds, Partly cloudy, Partly cloudy (날씨 정보가 1 \~ 4까지 정수형으로 들어가 있습니다.)    
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;          2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;          3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;         4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog   
temp - temperature in Celsius (온도 정보가 섭씨 기준으로 들어가 있습니다.)  
atemp - "feels like" temperature in Celsius (체감 온도가 섭씨 기준으로 들어가 있습니다.)   
humidity - relative humidity (상대 습도 값이 들어가 있습니다.)  
windspeed - wind speed (풍속 값이 들어가 있습니다.)  
casual - number of non-registered user rentals initiated (비회원 렌탈 대여수가 들어가 있습니다.)  
registered - number of registered user rentals initiated (회원 렌탈 대여수가 들어가 있습니다.)  
count - number of total rentals (총 대여수(= 비회원 렌탈 + 회원 렌탈)가 들어가 있습니다.)

루브릭 기준에 맞춰 RMSE를 도출했으나, R squared 값으로 판단해 보면 좋은 성능을 가지는 것은 아닌 것 같습니다. 어떻게 보면 데이터의 분포에 따라 모델의 성능도 크게 영향을 받을 수 밖에 없을 것 같다는 생각이 들었습니다. 실제로 규칙이 없는 데이터를 가지고 머신러닝 알고리즘으로 규칙을 찾는다라고 하면 찾을 수 있을 것인가 하는 의문이 들었습니다. 패턴이 있지만 패턴을 찾기 힘든 것을 머신러닝이 도와주는 것이지 원래 부터 패턴이 없는 데이터에서 패턴을 찾는다는 것이 가능한 것인가 하는 생각이 들었습니다. 따라서 머신러닝을 활용하는 관점을 전지 전능하다는 관점으로 접근하면 안된다는 생각이 들었습니다.   

본 프로젝트를 수행하면 어려웠던 점은 차원과 데이터 수를 맞추지 않아서 오류가 나는 부분이었습니다. 데이터를 일관성있게 관리하고 또한 차원에 대해서도 지속적으로 확인하면서 진행해야 오류를 줄일 수 있을 거라 생각됩니다.
