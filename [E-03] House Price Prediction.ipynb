### 3.4. Hyper Parameter Tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
from tqdm import tqdm
import time
```

```python
def myGrid(model, param, x, y):
    start = time.time()
    mygrid = GridSearchCV(model, param_grid = param, scoring = 'neg_mean_squared_error', n_jobs=5, verbose=1)
    mygrid.fit(x,y)
    params = mygrid.cv_results_['params']
    score = mygrid.cv_results_['mean_test_score']
    for i in tqdm(range(len(params))):
        params[i]['score'] = score[i]
    print(f'(time lapsed : {time.time() - start} sec.)',)
    return pd.DataFrame(params).sort_values(by='score', ascending=False)
```

### 3.4.4. Average Blending
상기 모델중 상대적으로 결과가 좋게 나온 XGBoost와 LightGBM 두 가지 모델의 결과를 산술 평균을 통해 Blending 모델을 만들겠습니다.

```python
models = [ {'model':xgb, 'name':'XGBoost'},
          {'model':lgb, 'name':'LightGBM'}]
```

```python
def AveragingBlending(models, x, y, test):
    for m in models : 
        m['model'].fit(x.values, y)
    
    pred = np.column_stack([
        m['model'].predict(test.values) for m in models
    ])
    return np.mean(pred, axis=1)
```

```python
y_pred = AveragingBlending(models, X_t, y_t,X_v)
mean_squared_error(np.expm1(y_v),np.expm1(y_pred)) **0.5
```

XGBoost와 LightGBM 두 가지 모델을 혼합하여 적용 결과 기존 최고 성능을 나타냈던 LightGBM 단독 모델보다 Loss값이 줄어 들었습니다.

```python
# lgb, xgb 하이퍼 파라미터 조정 후 재실행 결과
models = [ {'model':xgb2, 'name':'XGBoost'},
          {'model':lgb2, 'name':'LightGBM'}]

y_pred = AveragingBlending(models, X_t, y_t,X_v)
mean_squared_error(np.expm1(y_v),np.expm1(y_pred)) **0.5
```

### Make Submission

최종 제출은 가장 좋은 결과가 도출된 XGBoost와 LightGBM을 Average Bleding한 모델로 적용하도록 하겠습니다.

```python
def Prediction(models, df_test):
    pred = np.column_stack([
    m['model'].predict(df_test.values) for m in models
    ])
    return np.mean(pred, axis=1)
```
